  <h3 style="margin:0; color:#333;">Exercice 3 - Système différentiel : Modèle proie-prédateur de Lotka-Volterra</h3>
</div>

Le modèle de proie-prédateur permet de décrire la dynamique de populations d'un système biologique où interagissent un prédateur et sa proie. Il a été proposé de par Alfred James Lotka et Vito Volterra.

On note $x(t)$ et $y(t)$ les effectifs respectifs des proies et des prédateurs à l'instant $t$, le modèle s'écrit sous la forme du système d'équations différentielles suivantes :
$$\left\{\begin{array}{l} \dfrac{\mathrm{d}x(t)}{\mathrm{d}t} = x(t)\,(\alpha - \beta y(t)) \\[10pt] \dfrac{\mathrm{d}y(t)}{\mathrm{d}t} = y(t)\,(\delta x(t) - \gamma) \end{array}\right.$$

où :
<ul>
    <li>$\alpha$ est le taux de reproduction intrinsèque des proies ;</li>
    <li>$\beta$ est le taux de mortalité des proies due aux prédateurs ;</li>
    <li>$\delta$ est le taux de reproduction des prédateurs en fonction des proies mangées ;</li>
    <li>$\gamma$ est le taux de mortalité intrinsèque des prédateurs.</li>
</ul>

> 📝 **Question 17 —** Comment interpréter la première équation lorsque $\beta = 0$ ? et la seconde lorsque $\delta = 0$ ? Comment évoluent les populations concernées ?
